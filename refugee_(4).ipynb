```python
import folium
from folium.plugins import HeatMap

# Prepare data
population_by_origin = jordan_data.groupby("origin_location_code")["population"].sum().reset_index()

# Add coordinates
country_coords = {
    "SYR": [34.8021, 38.9968],  # Syria
    "IRQ": [33.2232, 43.6793],  # Iraq
    "YEM": [15.5527, 48.5164],  # Yemen
    "SDN": [12.8628, 30.2176],  # Sudan
    "SOM": [5.1521, 46.1996]    # Somalia
}
population_by_origin["lat"] = population_by_origin["origin_location_code"].map(lambda x: country_coords.get(x, [None, None])[0])
population_by_origin["lon"] = population_by_origin["origin_location_code"].map(lambda x: country_coords.get(x, [None, None])[1])

# Filter data
heatmap_data = population_by_origin.dropna(subset=["lat", "lon"])

# Create map
m = folium.Map(location=[20, 30], zoom_start=3)

# Prepare heatmap data
heat_data = [[row["lat"], row["lon"], max(row["population"], 1)] for index, row in heatmap_data.iterrows()]
HeatMap(heat_data).add_to(m)

# Save map
m.save("refugees_heatmap.html")
print("The map has been saved. Open 'refugees_heatmap.html' in your browser to view the map.")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is the loaded DataFrame
# Filter data for Jordan as the host country
jordan_data = df[df['asylum_location_code'] == 'JOR']

# Count occurrences of HRP and GHO for Jordan
jordan_hrp_counts = jordan_data['asylum_has_hrp'].value_counts()
jordan_gho_counts = jordan_data['asylum_in_gho'].value_counts()

# Create subplots for HRP and GHO
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# HRP plot for Jordan
jordan_hrp_counts.plot(kind='bar', ax=axs[0], color='skyblue')
axs[0].set_title('Jordan: Humanitarian Response Plan (HRP)')
axs[0].set_xticks(range(len(jordan_hrp_counts.index)))  # Dynamically set ticks
axs[0].set_xticklabels(jordan_hrp_counts.index.map({True: 'Has HRP', False: 'No HRP'}), rotation=0)
axs[0].set_xlabel("HRP Status")
axs[0].set_ylabel("Count")

# GHO plot for Jordan
jordan_gho_counts.plot(kind='bar', ax=axs[1], color='salmon')
axs[1].set_title('Jordan: Global Health Operations (GHO)')
axs[1].set_xticks(range(len(jordan_gho_counts.index)))  # Dynamically set ticks
axs[1].set_xticklabels(jordan_gho_counts.index.map({True: 'In GHO', False: 'Not in GHO'}), rotation=0)
axs[1].set_xlabel("GHO Status")
axs[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```
